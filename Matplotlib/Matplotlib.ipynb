## Functional method

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Object Oriented Method 

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #This parameter is the dimensions [left, bottom, width, height] 

axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure() #crete a empty canvas

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #add axes in the canvas
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
axes2.set_title('SMALLER PLOT')
axes1.set_title('LARGER PLOT')
```

```python
fig = plt.figure()
```

## SubPlots

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #tuple unpacking
fig.tight_layout()

# for current_axes in axes:
#     current_axes.plot(x,y)

axes[0].plot(x,y,'g-*')
axes[0].set_title("First Plot")

axes[1].plot(y,x,'r--')
axes[1].set_title("Second Plot")

plt.tight_layout() #helps to get rid of overlapping
```

## Plot appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color="purple", lw=1,linestyle='-', marker='o', markersize=10,
markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='green')
```
