# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.set_theme()
```

```python
sns.pairplot(iris,hue='species',palette='Paired')
```

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(sns.scatterplot)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(tips, row='time', col='smoker')
```

```python
g = sns.FacetGrid(tips, row='time', col='smoker')
g.map(plt.hist, 'total_bill')
```

```python
g = sns.FacetGrid(tips, row='time', col='smoker')
g.map(plt.scatter, 'total_bill', 'size')
```

```python
g = sns.FacetGrid(tips, row='time', col='smoker', hue='sex', palette='gnuplot2_r')
g.map(plt.scatter, 'total_bill', 'size').add_legend()
```
