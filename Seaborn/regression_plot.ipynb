```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_theme()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='Accent_r', markers=['o','v'],
scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', palette='BuGn')
```

```python
#Aspect will change the width while keeping the height constant
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', aspect=0.6, height=8)
```
